strategies and personalized business decisions.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy
```

```python
df=pd.read_csv('advance stage\\task2resources\\customer_segmentation.csv',encoding='latin-1')
df.head()
```

```python
df.describe()
```

```python
df=df.drop_duplicates()
df=df.dropna()
df_selected = df.select_dtypes(include=[np.number])
```

normalizing features

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_selected)
scaled_data
```

finding optimal k

```python
from sklearn.cluster import KMeans
inertia = []
for k in range(1, 11):
 km = KMeans(n_clusters=k)
 km.fit(scaled_data)
 inertia.append(km.inertia_)
```

applying k means 

```python
kmeans = KMeans(n_clusters=4)
df['Cluster'] = kmeans.fit_predict(scaled_data)
kmeans
```

```python
plt.scatter(df['Income'], df['Z_Revenue'], c=df['Cluster'])
```
